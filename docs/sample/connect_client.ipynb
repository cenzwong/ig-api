# Connect Client Sample

This notebook demonstrates how to connect to the IG API using the `IGClient`.

```python
from ig_api import IGClient

client = IGClient()
try:
    session_details = client.connect()
    print("Successfully connected!")
    print(f"Account ID: {session_details.get('account_id')}")
except Exception as e:
    print(f"Failed to connect: {e}")
```
